```python
import requests

tournament_id = 202   # Ekstraklasa
season_id = 61236     # Sezon 24/25

all_matches = []
ids = []

for offset in range(200):  # np. pierwsze 10 stron, aż się skończą mecze
    url = f"https://api.sofascore.com/api/v1/unique-tournament/{tournament_id}/season/{season_id}/events/last/{offset}"
    resp = requests.get(url).json()
    events = resp.get("events", [])


    if "events" not in resp or not resp['events']:
        print("no more games")
        print(len(all_matches))
        break
   
    for event in resp["events"]:
        temp = event['id']
        print(event["id"], event["homeTeam"]["name"], "-", event["awayTeam"]["name"])
        ids.append(temp)
    
    
    all_matches.extend(events)

print("Łącznie pobrano:", len(all_matches), "meczów")
```

```python
import ScraperFC as sfc
import pandas as pd

ss = sfc.Sofascore()

# Upewnij się, że masz listę kolumn do DataFrame
df = pd.DataFrame()
all_cols = set()
dfs = []

for match_id in ids:
    try:
        temp_df = ss.scrape_player_match_stats(match=int(match_id))

        # jeżeli coś się pobrało
        if temp_df is not None and not temp_df.empty:
            all_cols |= set(temp_df.columns)
            dfs.append(temp_df)
            print(f"Pobrano dane dla meczu {match_id} ({len(temp_df)} wierszy)")
        else:
            print(f"Brak danych dla meczu {match_id}")

    except Exception as e:
        print(f"Błąd przy meczu {match_id}: {e}")

df = pd.concat([d.reindex(columns=all_cols) for d in dfs], ignore_index=True)
print("Łącznie pobrano:", len(df), "wierszy")
```

```python
import pandas as pd

# 1. Usuń zduplikowane kolumny w każdej ramce
dfs_clean = [d.loc[:, ~d.columns.duplicated()] for d in dfs]

# 2. Zbierz wszystkie możliwe kolumny
all_cols = set()
for d in dfs_clean:
    all_cols |= set(d.columns)

# 3. Ujednolić kolumny i scalić
df = pd.concat([d.reindex(columns=all_cols) for d in dfs_clean], ignore_index=True)

print("Łącznie pobrano:", len(df), "wierszy, kolumny:", len(df.columns))
```
