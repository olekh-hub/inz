```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
liczba_iteracji = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 100000]
empiryczne_pi = [3.48, 3.27, 3.25, 3.1624, 3.1372, 3.1452, 3.14485, 3.144016, 3.143892]

wyniki = pd.DataFrame({'liczba_iteracji': liczba_iteracji, 'empiryczne_pi': empiryczne_pi})
wyniki['blad_bezwzgledny'] = abs(np.pi - wyniki['empiryczne_pi'])
```

```python
wyniki
```

```python
plt.figure(figsize=(14,7))
plt.plot(wyniki['liczba_iteracji'], wyniki['blad_bezwzgledny'])
plt.xlabel('Liczba iteracji')
plt.ylabel('Blad bezwzgledny')
plt.title('Zaleznosc bledu bezwzglednego od liczby iteracji')
plt.show()
```

```python
wyniki['stosunek_bledow'] = wyniki['blad_bezwzgledny'].shift(1) / wyniki['blad_bezwzgledny']
wyniki['stosunek_iteracji'] = wyniki['liczba_iteracji'] / wyniki['liczba_iteracji'].shift(1)
wyniki
```
